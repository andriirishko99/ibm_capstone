## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
query_result = %sql SELECT ConvertedComp FROM master
df = query_result.DataFrame()
df.hist()
```

### Box Plots


Plot a box plot of `Age.`


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# your code goes here
query_result = %sql SELECT Age FROM master
df = query_result.DataFrame()
plt.figure(figsize=(12,5))
sns.boxplot(df, orient = 'h')
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
query_result = %sql SELECT Age, WorkWeekHrs FROM master
df = query_result.DataFrame()

plt.figure(figsize=(10, 6))
plt.scatter(x=df['Age'], y=df['WorkWeekHrs'])

plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.title('Scatter Plot of Age vs. Work Week Hours')
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
query_result = %sql SELECT Age, CodeRevHrs, WorkWeekHrs FROM master
df = query_result.DataFrame()

plt.figure(figsize=(10, 6))
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], s=df['Age'], alpha=0.5)
# sns.scatterplot(x='WorkWeekHrs', y='CodeRevHrs', size='Age', data=df, alpha=0.5)

# Add labels and title
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.title('Bubble Plot of Work Week Hours vs. Code Review Hours')

# Show the plot
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
query_result = %sql SELECT DatabaseDesireNextYear AS Database, COUNT(DatabaseDesireNextYear) AS number FROM DatabaseDesireNextYear GROUP BY DatabaseDesireNextYear ORDER BY COUNT(DatabaseDesireNextYear) DESC LIMIT 5
df = query_result.DataFrame()

plt.figure(figsize=(8, 8))
plt.pie(df['number'], labels=df['Database'], autopct='%1.1f%%', startangle=90)

# Add a title
plt.title('Top 5 Databases')

# Show the plot
plt.show()
```

```python
# your code goes here
query_result = %sql SELECT LanguageDesireNextYear AS Language, COUNT(LanguageDesireNextYear) AS number FROM LanguageDesireNextYear GROUP BY LanguageDesireNextYear ORDER BY COUNT(LanguageDesireNextYear) DESC LIMIT 5
df = query_result.DataFrame()

plt.figure(figsize=(8, 8))
plt.pie(df['number'], labels=df['Language'], autopct='%1.1f%%', startangle=90)

# Add a title
plt.title('Top 5 Languages')

# Show the plot
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
query_result = %sql SELECT WorkWeekHrs, CodeRevHrs FROM master WHERE Age > 30 AND Age < 35
df = query_result.DataFrame()
median_work = df['WorkWeekHrs'].median()
median_code = df['CodeRevHrs'].median()


plt.figure(figsize=(4, 6))
plt.bar('Age Group 30-35', median_work, color='blue')

# Plot the second bar for median "CodeRevHrs" on top of the first bar
plt.bar('Age Group 30-35', median_code, color='orange', bottom=median_work)

# Add labels and title
plt.ylabel('Median Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs for Ages 30 to 35')

# Show the plot
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
query_result = %sql SELECT ConvertedComp, Age FROM master WHERE Age > 45 AND Age < 60
df = query_result.DataFrame()
new_df = df.groupby(by=df['Age']).mean()

plt.figure(figsize=(10, 6))
plt.plot(new_df['ConvertedComp'], marker='o', linestyle='-')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('ConvertedComp')
plt.title('ConvertedComp median from 45 to 60 ')

# Show the plot
plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
query_result = %sql SELECT MainBranch, COUNT(MainBranch) AS Frequency FROM master GROUP BY MainBranch
df = query_result.DataFrame()

plt.figure(figsize=(10, 6))
sns.barplot(x = 'Frequency', y = 'MainBranch', data= df, orient = 'h')

# Add labels and title
plt.xlabel('Number of Respondents')
plt.ylabel('Main Branch')
plt.title('Distribution of Main Branches')

# Show the plot
plt.show()
```
